```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
os.chdir("E:/EBAC/Material/M21")

df = pd.read_csv("recursos_humanos.csv")
df
```

```python
df.drop(['number_project'], axis = 1, inplace=True)
```

```python
df_codificado = pd.get_dummies(df, columns=['sales', 'salary'], drop_first=True)
df_codificado
```

```python
# Convertir valores boleanos (True/False) a enteros (1/0)

bool_columns = df_codificado.select_dtypes(include=['bool']).columns
df_codificado[bool_columns] = df_codificado[bool_columns].astype(int)
df_codificado
```

Para equilibrar el caso, debemos aplicar normalizacion a los datos debido a los diferentes rangos y magnitudes que se manejan en las diferentes categorias

```python
# Primero separar el factor objetivo del resto de la informacion
y = df_codificado.left.values
x_data = df_codificado.drop(["left"], axis = 1)
```

```python
# CreaciÃ³n de grupos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_data, y, test_size=0.2, random_state=4)
print("Grupo de entrenamiento", X_train.shape, y_train.shape)
print("Grupo de prueba", X_test.shape, y_test.shape)
```

```python
from sklearn import svm

# Radial Basis function (RBF)
clf = svm.SVC(kernel="rbf")
clf.fit(X_train, y_train)

yhat= clf.predict(X_test)
yhat[0:5]
```

```python
from sklearn.metrics import confusion_matrix
y_pred = clf.predict(x_data)
cm = confusion_matrix(y_test, yhat)
cm
```

```python
# Visualizacion de la matriz de confusiÃ³n
import seaborn as sns
f, ax = plt.subplots(figsize=(5,5))

sns.heatmap(cm, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("y pronosticada")
plt.ylabel("y real")
plt.show()
```

```python
from sklearn.metrics import classification_report
cnf_matrix = confusion_matrix(y_test, yhat, labels=[0,1])
print(classification_report(y_test, yhat))
```

La utilizacion del kernel Radial Basis function (RBF) no resulta ser muy optimo para este caso ya que aunque no tuvo errores con el personal que no dejo la compaÃ±ia, hace una pobre estimacion de los que si se fueron que es el factor mas importante de este estudio

```python
# Lineal

clf = svm.SVC(kernel="linear")
clf.fit(X_train, y_train)

yhat= clf.predict(X_test)

y_pred = clf.predict(x_data)
cm = confusion_matrix(y_test, yhat)
cm
```

```python
f, ax = plt.subplots(figsize=(5,5))

sns.heatmap(cm, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("y pronosticada")
plt.ylabel("y real")
plt.show()
```

```python
cnf_matrix = confusion_matrix(y_test, yhat, labels=[0,1])

print(classification_report(y_test, yhat))
```

Aunque este kernel (linear) mejora contra el anterior, sigue realizando un pronostico pobre sobre el factor importante y el pronostico acertado de los empleados que se fueron sigue siendo opacado por los errores que tuvo, convirtiendolo en una herramienta poco util para este caso

```python
# Polinomial

clf = svm.SVC(kernel="poly")
clf.fit(X_train, y_train)

yhat= clf.predict(X_test)

y_pred = clf.predict(x_data)
cm = confusion_matrix(y_test, yhat)
cm
```

```python
# Sigmoidal

clf = svm.SVC(kernel="sigmoid")
clf.fit(X_train, y_train)

yhat= clf.predict(X_test)

y_pred = clf.predict(x_data)
cm = confusion_matrix(y_test, yhat)
cm
```

```python
f, ax = plt.subplots(figsize=(5,5))

sns.heatmap(cm, annot=True, linewidths=0.5, linecolor="red", fmt=".0f", ax=ax)
plt.xlabel("y pronosticada")
plt.ylabel("y real")
plt.show()
```

El kernel Sigmoidal parece ser el mas apropiado para este caso debido a tener el balance mas optimo de los casos anteriores, es el modelo que obtuvo un mayor numero del factor que resulta mas importante determinar que en este caso es los empleados que se fueron de la compaÃ±ia y aunque no es optimo, o confiable, los errores obtenidos se balancean entre si, lo que nos permite usar este metodo para obtener cierta informacion util pero la utilizacion de otro metodo u optimizacion de los datos seria recomendado.
